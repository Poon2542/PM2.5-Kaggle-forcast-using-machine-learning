```python
traffic2013['ปี/เดือน/วัน'] = np.nan
traffic2013['ชั่วโมง'] = np.nan
```

```python
traffic2013
```

```python
def assignDatetime(tf_dataframe):
  for i in range(len(tf_dataframe)):
    dt = str(tf_dataframe.iloc[i]["datetime"])
    ymd = dt[2:4]+dt[5:7]+dt[8:10]
    hour = int(dt[11:13])*100
    if hour==0: hour =2400

    tf_dataframe.at[i,'ปี/เดือน/วัน'] = ymd
    tf_dataframe.at[i,'ชั่วโมง'] = hour
```

```python
assignDatetime(traffic2013)
```

```python
traffic2013
```

```python
traffic2013_h = traffic2013.groupby(["ปี/เดือน/วัน","ชั่วโมง"]).mean(numeric_only=True)
```

```python
traffic2013_h
```

```python
traffic2013_h.to_csv("traffic2013_h.csv")
```

```python
traffic_list = [traffic2014,traffic2015,traffic2016,traffic2017,traffic2018,traffic2019,traffic2020,traffic2021,traffic2022]
filename_list = ["traffic2014","traffic2015","traffic2016","traffic2017","traffic2018","traffic2019","traffic2020","traffic2021","traffic2022"]
```
