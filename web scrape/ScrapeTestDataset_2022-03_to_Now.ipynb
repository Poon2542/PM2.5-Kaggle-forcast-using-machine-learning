```python
def urlRender(station):
  startdate = "2020-10-01"
  starttime = "00"
  url ="http://air4thai.com/forweb/getHistoryData.php?stationID="+station+"&param=WS,WD,TEMP,RH,PM10,PM25&type=hr&sdate="+startdate+"&edate="+datenow+"&stime="+starttime+"&etime="+timenow
  return url
```

# Join traffic

```python
result_nn = result[result["PM2.5"].isnull() == False]
result_nn
```

```python
tf = pd.read_csv('/content/drive/MyDrive/DataScience_finalproject/Traffics/traffic_hour/traffic_hour.csv')
```

```python
traffic = tf[["ปี/เดือน/วัน","ชั่วโมง","index"]]
traffic
```

```python
traffic1 = pd.DataFrame({'ปี/เดือน/วัน':[],'ชั่วโมง':[],'index':[]})
```

```python
for i in range(len(traffic)):
    ymd = traffic.iloc[i]["ปี/เดือน/วัน"]
    hour = int(traffic.iloc[i]['ชั่วโมง'])/100
    if hour==24: hour =0

    traffic.at[i,'ปี/เดือน/วัน'] = float(ymd)
    traffic.at[i,'ชั่วโมง'] = float(hour)

    row = {'ปี/เดือน/วัน':ymd,'ชั่วโมง':hour,'index':traffic.iloc[i]["index"]}
    traffic1 = traffic1.append(row, ignore_index = True)
```

```python
traffic1
```

```python
result_nn = result_nn.join(traffic1.set_index(['ปี/เดือน/วัน','ชั่วโมง']), on=['ปี/เดือน/วัน','ชั่วโมง'])
```
